# Visualizing Toronto Neighborhoods

Load all the packages:

```python
import pandas as pd
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
!conda install -c conda-forge folium=0.5.0
import folium
from geopy.geocoders import Nominatim

print('Libraries imported.')
```

Load the dataframe with the neighborhoods and their longitude and latitude from the last notebook:

```python
my_file = project.get_file("df_longlat.csv")

# Read the CSV data file from the object storage into a pandas DataFrame
my_file.seek(0)
df_ll=pd.read_csv(my_file)
df_ll.head()
```

Create a map of Toronto with the postal codes marked.

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="toronto_exp")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, postcode in zip(df_ll["Latitude"], df_ll["Longitude"], df_ll["Postal_Code"]):
    label = '{}'.format(postcode)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

Let's take a closer look at downtown Toronto:

```python
downtown_data = df_ll[df_ll['Borough'] == 'Downtown Toronto'].reset_index(drop=True)
downtown_data.head()
```

```python
map_downtown = folium.Map(location=[latitude, longitude], zoom_start=14)

# add markers to map
for lat, lng, postcode in zip(downtown_data["Latitude"], downtown_data["Longitude"], downtown_data["Postal_Code"]):
    label = '{}'.format(postcode)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_downtown)  
    
map_downtown
```
